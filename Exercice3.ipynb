```python
#Charger les packages
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
#2: Changer les variables:
for col in ['prof', 'duree', 'discip', 'ecole', 'separation', 'juge.enfant', 'place', 'abus', 'grav.cons', 'dep.cons', 'ago.cons','ptsd.cons','alc.cons', 'subst.cons', 'scz.cons', 'char', 'rs', 'ed', 'dr','suicide.hr','suicide.past']:
    data[col]=data[col].astype('category')
data.dtypes
```

```python
#3: Moyenne, variance et ecart-type
#Pour l'age:
moyenne_age=data['age'].mean()
variance_age=data['age'].var()
sigma_age=data['age'].std()
print(f"La moyenne de la variable age est: {moyenne_age}.\nSa variance est: {variance_age}.\nSon ecart-type est: {sigma_age}")
print("------------------------------------------------------------")
#Pour n.enfant:
moyenne_enfant=data['n.enfant'].mean()
variance_enfant=data['n.enfant'].var()
sigma_enfant=data['n.enfant'].std()
print(f"La moyenne de la variable n.enfant est: {moyenne_enfant}.\nSa variance est: {variance_enfant}.\nSon ecart-type est: {sigma_enfant}")
print("------------------------------------------------------------")
#Pour n.fratrie:
moyenne_fratrie=data['n.fratrie'].mean()
variance_fratrie=data['n.fratrie'].var()
sigma_fratrie=data['n.fratrie'].std()
print(f"La moyenne de la variable n.fratrie est: {moyenne_fratrie}.\nSa variance est: {variance_fratrie}.\nSon ecart-type est: {sigma_fratrie}")
print("-------------------------------------------------------------")
#Pour dur.interv:
moyenne_i=data['dur.interv'].mean()
variance_i=data['dur.interv'].var()
sigma_i=data['dur.interv'].std()
print(f"La moyenne de la variable dur.interv est: {moyenne_i}.\nSa variance est: {variance_i}.\nSon ecart-type est: {sigma_i}")
```

```python
#Donnons les 3 premiers quantiles pour la variable age
q1=data['age'].quantile(0.25)
q2=data['age'].quantile(0.50)
q3=data['age'].quantile(0.75)
print(f"q1={q1}; q2={q2}; q3={q3}")
```

```python
#4: Boxplot de la variable age
#Les valeurs nan ne permettent pas de tracer le boxplot donc on les remplace par la moyenne dans ce cas
a=data['age'][::]
a[np.isnan(a)]=a.mean()
plt.figure(figsize=(10,4))
plt.boxplot(a)
plt.title('Boxplot des Ã¢ges')
plt.savefig('Boxplot.png')
plt.show()
```

```python
#Diagramme circulaire de la variable profession
labels="nan", "employe", "ouvrier", "agriculteur", "cadre ", "prof.intermediaire", "artisan", "autre", "sans emploi"
sizes=dc
colors=['black', 'greenyellow', 'coral', 'aqua', 'cadetblue', "pink", "silver", "tan", "khaki"]
plt.figure(figsize=(15,7))
plt.pie(sizes, labels=labels, colors=colors, shadow=False, startangle=180)
plt.axis('equal')
plt.title('Diagramme Circulaire des professions')
plt.savefig('Cammembert_des_ages.png')
plt.show()
```

```python
#13, 14: DiscrÃ©disation de la variable Ã¢ge et frÃ©quence des modalitÃ©s
import sys
min_age=int(data['age'].min())
max_age=int(data['age'].max())
Q1=int(data['age'].quantile(0.25))
Q2=int(data['age'].quantile(0.5))
Q3=int(data['age'].quantile(0.75))
data['age_classe']=pd.qcut(data['age'], 4, labels=[f'[{min_age}, {Q1}]', f']{Q1}, {Q2}]', f']{Q2}, {Q3}]', f']{Q3}, {max_age}]'])
total=data['age_classe'].count()
data1=data['age_classe'].where(data['age_classe']==f'[{min_age}, {Q1}]')
data2=data['age_classe'].where(data['age_classe']==f']{Q1}, {Q2}]')
data3=data['age_classe'].where(data['age_classe']==f']{Q2}, {Q3}]')
data4=data['age_classe'].where(data['age_classe']==f']{Q3}, {max_age}]')
print('Les frÃ©quences des modalitÃ©s de la variable classe sont:\n ')
print(f'[min(age), Q1] = [{min_age}, {Q1}] = %.3f'%(data1.count()/total *100)+'%')
print(f']Q1, Q2] = ]{Q1}, {Q2}] = %.3f'%(data2.count()/total *100)+'%')
print(f']Q2, Q3] = ]{Q2}, {Q3}] = %.3f'%(data3.count()/total *100)+'%')
print(f']Q3, max(age)] = ]{Q3}, {max_age}] = %.3f'%(data4.count()/total *100)+'%\n\n\n')
```
